```python
# Zadanie 4 Utwórz tablicę numpy o wymiarach 5x5 z kolejnymi wartościami całkowitoliczbowymi od 1 do 25 i następnie wyświetl:
tab = np.arange(1, 26).reshape(5, 5)
print(tab)
print()
print(f"Ostatnia wartość z tablicy: {tab[4,4]}")
print(f"Druga kolumna: {tab[:,1]}")
print(f"Trzeci wiersz: {tab[2,:]}")
print(f"Główna przekątna: {tab.diagonal()}")
print(f"""tablicę, która zawiera wartości 9,10,14 oraz 15:
{tab[1:3,3:5]}""")
```

```python
# Zadanie 5 (1/2)
import numpy as np
import matplotlib.pyplot as plt

x = np.ones((3, 3, 3))

calosc = (x < 2)

#calosc[x, z, y]

calosc[:, :, [0,2]] = False


display(calosc)

ax = plt.figure().add_subplot(projection='3d')
ax.voxels(calosc, facecolors='orange', edgecolor='k')

plt.show()
```

```python
# Zadanie 5 (2/2)
import numpy as np
import matplotlib.pyplot as plt

x = np.ones((3, 3, 3))

calosc = (x < 2)

calosc[1, [0,2], :] = False
calosc[1, :, [0,2]] = False
calosc[1, 1, 1] = False

display(calosc)

ax = plt.figure().add_subplot(projection='3d')
ax.voxels(calosc, facecolors='orange', edgecolor='k')

plt.show()
```

```python
N = 10000

def test_for_loop():
    lista = list(range(1, N+1))
    start_time = time.time()
    for idx in range(len(lista)):
        lista[idx] = lista[idx] * 2
    return time.time() - start_time

def test_list_comprehension():
    lista = list(range(1, N+1))
    start_time = time.time()
    lista = [elem * 2 for elem in lista]
    return time.time() - start_time

def test_numpy():
    arr = np.arange(1, N+1)
    start_time = time.time()
    arr = arr * 2
    return time.time() - start_time

# Wykonanie testów
time_for = test_for_loop()
time_comp = test_list_comprehension()
time_numpy = test_numpy()

# Wyświetlenie wyników
print(f"Czas wykonania pętli for: {time_for:.6f} s")
print(f"Czas wykonania list comprehension: {time_comp:.6f} s")
print(f"Czas wykonania NumPy: {time_numpy:.6f} s")
```
